```python
import pandas as pd

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# CARGAR CONVERSACIONES
conversaciones = pd.read_csv(
    "../data/processed/conversaciones_clean.csv"
)
```

```python
# CARGAR MODELO
model = SentenceTransformer(
    'all-MiniLM-L6-v2'
)
```

```python
# LIMPIEZA
conversaciones["input"] = (
    conversaciones["input"]
    .fillna("")
    .astype(str)
)
```

```python
# CREAR EMBEDDINGS
embeddings = model.encode(
    conversaciones["input"].tolist(),
    show_progress_bar=True
)
```

```python
# PRUEBAAAAAA
query = "Hola!"

query_embedding = model.encode([query])
```

```python
# ENCONTRAR SIMILITUD
similarities = cosine_similarity(
    query_embedding,
    embeddings
)
```

```python
# ENCONTRAR RESPUESTA PARECIDA
best_match = similarities.argmax()

print("Pregunta encontrada:")
print(conversaciones.iloc[best_match]["input"])

print("\nRespuesta:")
print(conversaciones.iloc[best_match]["output"])
```

```python
replacements = {
    "Hey Banco": "NovaBank",
    "Hey": "Nova",
    "Hey Pro": "Nova+",
    "Cuenta Hey": "Cuenta Nova",
    "Hey Negocios": "Nova Business"
}
```

```python
for old, new in replacements.items():

    conversaciones["input"] = (
        conversaciones["input"]
        .str.replace(old, new, regex=False)
    )

    conversaciones["output"] = (
        conversaciones["output"]
        .str.replace(old, new, regex=False)
    )
```

```python
embeddings = model.encode(
    conversaciones["input"].tolist(),
    show_progress_bar=True
)
```
