```python
import pandas as pd
import numpy as np
```

```python
clientes = pd.read_csv(
    "../data/processed/clientes_clean.csv"
)

productos = pd.read_csv(
    "../data/processed/productos_clean.csv"
)

transacciones = pd.read_csv(
    "../data/processed/transacciones_clean.csv"
)

conversaciones = pd.read_csv(
    "../data/processed/conversaciones_clean.csv"
)
```

```python
transacciones["fecha_hora"] = pd.to_datetime(
    transacciones["fecha_hora"]
)
```

```python
# FEATURING ENGINE
```

```python
# FEATURE 1 - GASTO TOTAL
gasto_total = (
    transacciones.groupby("user_id")["monto"]
    .sum()
)
```

```python
# FEATURE 2 - GASTO PROMEDIO
gasto_promedio = (
    transacciones.groupby("user_id")["monto"]
    .mean()
)
```

```python
# FEATURE 3 — NÚMERO DE TRANSACCIONES
num_transacciones = (
    transacciones.groupby("user_id")
    .size()
)
```

```python
# FEATURE 4 — CATEGORÍA FAVORITA
categoria_favorita = (
    transacciones.groupby("user_id")["categoria_mcc"]
    .agg(lambda x: x.value_counts().index[0])
)
```

```python
# FEATURE 5 — CASHBACK TOTAL
cashback_total = (
    transacciones.groupby("user_id")["cashback_generado"]
    .sum()
)
```

```python
# FEATURE 6 — TRANSACCIONES INTERNACIONALES
internacionales = (
    transacciones.groupby("user_id")["es_internacional"]
    .sum()
)
```

```python
# FEATURE 7 — TRANSACCIONES ATÍPICAS
atipicas = (
    transacciones.groupby("user_id")["patron_uso_atipico"]
    .sum()
)
```

```python
# DATA SET FINAAAAAL
```

```python
features = pd.DataFrame({
    "gasto_total": gasto_total,
    "gasto_promedio": gasto_promedio,
    "num_transacciones": num_transacciones,
    "categoria_favorita": categoria_favorita,
    "cashback_total": cashback_total,
    "transacciones_internacionales": internacionales,
    "transacciones_atipicas": atipicas
})
```
